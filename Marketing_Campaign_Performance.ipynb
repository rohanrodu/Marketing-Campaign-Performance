```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV file (same folder)
df = pd.read_csv("F:/marketing_campaign.csv")

# Preview data
df.head(30)
```

```python

# Check structure
df.dtypes
df.isnull().sum()
df.shape
df.describe()

# Convert date column
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
```

```python
# Check duplicates
num_duplicates = df.duplicated().sum()
print("Duplicate rows:", num_duplicates)
```

```python
# Remove duplicates
cleaned_df = df.drop_duplicates()

# Save cleaned file
cleaned_df.to_csv("marketing_campaign_cleaned.csv", index=False)

#Cleaned file will saved in the same folder where the code file is located
```
